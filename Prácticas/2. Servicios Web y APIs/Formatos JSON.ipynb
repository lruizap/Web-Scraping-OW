# Formatos JSON

```python
# Vamos a leer un fichero JSON para hacer lo mismo que en el ejercicio anterior

# importamos librerias 

import json 

# usamos la función para leer el fichero "r" => read
f = open("data.json", "r")

info = json.loads(f.read())
```

```python
info
```

```python
print("Número de presidentes", len(info))
```

```python
for data in info:
    print('\r')
    print('Nombre:', data['nm'])
    print('Partido:', data['pp'])
    print('Años:', data['tm'])
```
