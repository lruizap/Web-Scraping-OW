```python
import urllib
from bs4 import BeautifulSoup

url = 'http://www-math.mit.edu/~gs/'

# Vamos a extraer los enlaces de la web
# 1º Cargamos el HTML
html = urllib.request.urlopen(url)
# 2º Covertimos el html en un objeto BeautifulSoup
soup = BeautifulSoup(html)
# 3º Extraemos las etiquetas
tags = soup('a')

print('Enlaces de la página principal: \r\n')
for tag in tags:
    print(tag.contents[0], tag.get('href'))
```

```python
# Las etiquetas se nos guardan en una lista
len(tags)
# Tenemos 32 etiquetas. Tendremos que acceder a cada uno de ellos y extraer los enlaces de estos
```

```python
print('Enlaces de las páginas secundarias')

for tag in tags:
    newUrl= tag.get('href')
    print('* Accediendo a los enlaces dentro de la página', newUrl)
    print()
    try:
        if newUrl[0:4] == 'http':
            # Si la url comienza por 'http', podemos acceder a ella sin problemas
            html2 = urllib.request.urlopen(newUrl)
        else:
            # Estas url son relativas, por tanto, para obtener la página tendremos que concatenar
            # la url de la página principal con la nueva
            html2 = urllib.request.urlopen(url+newUrl)
            
        soup2 = BeautifulSoup(html2)
        newTags = soup2('a')
        
        # Debemos indicarle al programa lo que sucede si dentro de la etiqueta hay nuevos enlaces
        # Si no encuentra enlaces deberemos indicarlo
        
        if len(newTags) > 0:
            print(len(newTags), ' enlaces:')
            for newTag in newTags:
                print(newTag.contents[0], newTag.get('href'))
            print()
        else: 
            print('No hay más enlaces')
            print()
    except:
        print('Algo ha fallado')
        print()
```
