```python
# Importamos las librerías necesarias para trabajar con html
import urllib
from bs4 import BeautifulSoup

# Ponemos la url donde se encuentra el discurso
url= 'https://elpais.com/internacional/2018/01/31/actualidad/1517387619_036241.html'

# extraemos el html
html = urllib.request.urlopen(url)

# Ahora transformamos el html en un objeto Beautiful Soup
soup = BeautifulSoup(html)

# Tendremos que investigar dentro del html donde se encuentra el discurso para
# que nuestro programa busque
# Tenemos que extrar las etiquetas indicadas
tags = soup('p')
```

```python
tags
```

```python
# Ahora recorreremos las etiquetas para añadirla a una etiqueta denominada discurso

discurso = ''

for tag in tags:
    # Si no tiene atributos lo añade
    if(len(tag.attrs))!=0:
        a = tag.contents[0]
        discurso = discurso + a

print(discurso)
```

```python
tags[0].contents[0]
```

```python
# Ahora contaremos las palabras

# Creamos diccionario
contadores = dict()

# Hacemos una modificación para quitar comas, espacios en blanco...
discurso = discurso.replace(",",'').replace('.','').replace('?','')
discurso = discurso.replace('¿','').replace('"','').replace('!','')
discurso = discurso.replace('-','').replace("'",'').replace('¡','')

# También tendremos que poner todo el texto en minúsculas para que no
# nos cuente la misma palabra dos veces

discurso = discurso.lower()

# Ahora añadismos las palabras al array
palabras = discurso.split()
```

```python
palabras
```

```python
for palabra in palabras:
    if len(palabra)> 3:
        # La función get lo que hace es ver cuántas veces hay una palabra en
        # El diccioario y suma
        contadores[palabra] = contadores.get(palabra,0) + 1
```

```python
contadores
```

```python
# Procedemos a pasar los datos a una tabla
import pandas
# esto nos permite pasar los datos del diccionario a tabla y ordenarlos
pandas.DataFrame(list(contadores.items()),columns=['palabra','contador']).sort_values('contador',ascending=False)
```
