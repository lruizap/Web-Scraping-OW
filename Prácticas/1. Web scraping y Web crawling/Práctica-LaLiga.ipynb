# Vamos a sacar datos de la página de LaLiga

```python
# Importamos las librerías

import urllib
from bs4 import BeautifulSoup
url = 'https://www.laliga.es'

# Tendremos que encontrar el contenido en el html y extraerlo
html = urllib.request.urlopen(url)
soup = BeautifulSoup(html)

# Vamos a filtrar las etiquetas por la clase
tags = soup.find_all("div", class_="styled__NextMatchContainer-sc-1qml4wk-0 cBta-dM")
```

```python
tags
```

```python
hashtags = list()

for tag in tags: 
    hashtags.append(tag.contents[0])
```
